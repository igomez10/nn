```python
import torch

tensor1 = torch.tensor([[1.0, 2.0]])
tensor2 = torch.tensor([[5.0, 6.0]])

print("tensor1:", tensor1)
print("tensor1 shape:", tensor1.shape)
print("tensor2:", tensor2)

# embedding 1
embedding1 = torch.nn.Embedding(10, 2)

# embed tensor 1
tensor3 = torch.tensor([1, 0, 2, 4])
embedded1 = embedding1(tensor3)
assert embedded1.shape == (4, 2)
# because tensor3 has 4 elements and embedding1 outputs 2-dimensional vectors
# so the output shape is (4, 2)


print("embedded1:", embedded1)
```
