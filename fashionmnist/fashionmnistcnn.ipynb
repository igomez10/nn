```python
import matplotlib.pyplot as plt
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
from helper import train_loop
from helper import fashionClassifierCNN
from helper import get_data
from helper import labels_map

device = (
    torch.accelerator.current_accelerator().type
    if torch.accelerator.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
training_data, test_data = get_data()

# split train dataset into train+validation
training_data, validation_data = torch.utils.data.random_split(training_data, [50000, 10000])

batch_size = 128
train_dataloader = DataLoader(training_data, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=batch_size, shuffle=False)
```

```python
model = fashionClassifierCNN(num_classes=10, in_channels=1).to(device)

optimizer = torch.optim.Adam(model.parameters())
train_loop(epochs=100, dataloader=train_dataloader, model=model, lossFn=nn.CrossEntropyLoss(), optimizer=optimizer, device=device)
```
