```python
from random import randint

# process the data
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
```

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(0)

    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)


for i in range(1000):
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(0)

    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)


for i in train_samples:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)


# shuffle
train_labels, train_samples = shuffle(train_labels, train_samples, random_state=42)

# scale the data


scaler = MinMaxScaler(feature_range=(0, 1))

# the reshape is needed because the scaler expects a 2D array
# -1, 1 means that we want to keep the number of rows and have 1 column
# the -1 is a placeholder for the number of rows, which will be inferred from the data
# the 1 means that we want to have 1 column, which is the number of features
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1, 1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# Simple tf.keras sequential model

```python
import tensorflow as tf
from keras.layers import Dense

# from tensorflow import keras
from keras.models import Sequential
```

```python
physical_devices = tf.config.experimental.list_physical_devices("CPU")
print("Num CPUs Available: ", len(physical_devices))
# tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
model = Sequential(
    [
        # Input(shape=(1,)),
        Dense(units=16, input_shape=(1,), activation="relu"),
        Dense(units=32, activation="relu"),
        Dense(units=2, activation="softmax"),
    ]
)

model.summary()
```

```python
import mlflow

mlflow.set_tracking_uri("https://mlflow-1055133852837.us-central1.run.app/")
mlflow.set_experiment("keras-tutorial")
mlflow.tensorflow.autolog()
```

```python
# compile the model: this basically configures the model
from keras.losses import sparse_categorical_crossentropy

model.compile(
    loss=sparse_categorical_crossentropy,
    metrics=["accuracy"],
)

with mlflow.start_run():
    model.fit(
        x=scaled_train_samples,
        y=train_labels,
        batch_size=10,
        validation_split=0.1,
        epochs=20,
        shuffle=True,
        verbose=2,
    )
```
