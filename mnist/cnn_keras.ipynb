```python
import keras
from keras import layers
```

```python
# Load data
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize and add channel dimension
max_pixel_value = 255.0
x_train = x_train.astype("float32") / max_pixel_value
x_test = x_test.astype("float32") / max_pixel_value
x_train = x_train[..., None]  # (N, 28, 28, 1)
x_test = x_test[..., None]


image_dimension = 28
kernel_size = 3
filter_count = 32
model = keras.Sequential(
    [
        # First block of convolutional layers
        layers.Input(shape=(image_dimension, image_dimension, 1)),
        # Add a convolutional layer with ReLU activation
        layers.Conv2D(filter_count, kernel_size, padding="same", activation="relu"),
        # Add a second convolutional layer to increase the model's capacity
        layers.Conv2D(filter_count, kernel_size, padding="same", activation="relu"),
        # Add a max pooling layer to reduce spatial dimensions
        layers.MaxPooling2D(),
        # Increase the number of filters in the second block
        layers.Conv2D(filter_count * 2, kernel_size, padding="same", activation="relu"),
        # Add another convolutional layer to further increase capacity
        layers.Conv2D(filter_count * 2, kernel_size, padding="same", activation="relu"),
        # Add another max pooling layer to further reduce spatial dimensions
        layers.MaxPooling2D(),
        # Flatten the output and add fully connected layers for classification
        layers.Flatten(),
        # Add a fully connected layer with ReLU activation to learn complex patterns
        layers.Dense(128, activation="relu"),
        # Add the output layer with softmax activation for multi-class classification
        layers.Dense(10, activation="softmax"),
    ]
)

model.compile(
    optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"]
)

model.fit(x_train, y_train, epochs=3, batch_size=64, validation_split=0.1)
model.evaluate(x_test, y_test)
```
