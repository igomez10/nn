```python
import torch

# set device to GPU if available, otherwise use CPU
device = (
    torch.accelerator.current_accelerator().type
    if torch.accelerator.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
# create tensor with 100 rows and 10 columns, filled with random values
X = torch.randn(10000000000, 1)
X.to(device)  # move tensor to GPU if available
```
