```python
import torch
from torch import nn

dim_size = 3
arpit_embedding = nn.Embedding(1, dim_size)
# set weights of the embedding layer to value [0.5, 0.1, 0.2]
arpit_embedding.weight.data = torch.tensor([[0.5, 0.1, 0.2]])
print(arpit_embedding.weight)


msds_students = nn.Embedding(79, dim_size)
# print(msds_students.weight)


ignacio_id = 10
yancy_id = 20

ts = torch.LongTensor([[ignacio_id, yancy_id, ignacio_id]])
some_embeddings = msds_students(ts)
print(some_embeddings)
print(some_embeddings.shape)
```
